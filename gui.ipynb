```python
''' main project'''
from tkinter import *
from random import randint
from tkinter.ttk import Progressbar
import tkinter.messagebox
root=Tk()
root.title("CRYPTOGRAPHER")
root.geometry("900x600+0+0")
root.configure(background='#ccffff')

label=Label(root,text="WELCOME TO CRYPTOGRAPHER",font=("arial",30,"bold"),fg="black")
label.grid()
photo=PhotoImage(file="password.png")
labelt=Label(root,image=photo)
labelt.grid()
pic=PhotoImage(file="hck.png")
labelx=Label(root,image=pic)
labelx.grid(row=1,column=1)

ALPHABET = 'abcdefghijklmnopqrstuvwxyz'
#if the user clicks on 1
def generate_otp(sheets,length):
    for sheet in range(sheets):
        with open("otp" + str(sheet) + ".txt","w") as f:
            for i in range(length):
                f.write(str(randint(0,26))+"\n")
    
def load_sheet(filename):
	with open(filename, "r") as f:
		contents = f.read().splitlines()
	return contents
def get_plaintext():
    label4=Label(root,text='Please type your message ',font=("arial",10,"bold"),fg="steelblue")
    mess=StringVar()
    mess.set('jon snow is daenerys nephew')
    entry_4=Entry(root,textvariable=mess,width=27,show='*')
    label4.grid(row=10)
    entry_4.grid(row=10,column=1)
    plaintext=str(mess.get())
    return plaintext.lower()
def encrypt(plaintext, sheet):
	ciphertext = ''
	for position, character in enumerate(plaintext):
		if character not in ALPHABET:
			ciphertext += character
		else:
			encrypted = (ALPHABET.index(character) + int(sheet[position])) % 26
			ciphertext += ALPHABET[encrypted]
	return ciphertext
def save_file(filename, data):
	with open(filename, 'w') as f:
		f.write(data)


def otp():
        
        label1=Label(root,text='how many one time pads you would like to generate?',font=("arial",10,"bold"),fg="steelblue")
        label2=Label(root,text='What will be your maximum message length? ',font=("arial",10,"bold"),fg="steelblue")
        sheet=IntVar()
        sheet.set(4)
        entry_1=Entry(root,textvariable=sheet)
        length=IntVar()
        length.set(27)
        entry_2=Entry(root,textvariable=length)
        label1.grid(row=7)
        label2.grid(row=8)
        entry_1.grid(row=7,column=1)
        entry_2.grid(row=8,column=1)
        
        tc=str(sheet.get())
        td=str(length.get())
        t=int(tc)
        d=int(td)
        generate_otp(t,d)
        
 #if the user clicks on 2        
def encry():
    label3=Label(root,text='type in the file name of the OTP you want to use',font=("arial",10,"bold"),fg="steelblue")
    fname=StringVar()
    fname.set('otp0.txt')
    entry_3=Entry(root,textvariable=fname)
    label3.grid(row=9)
    entry_3.grid(row=9,column=1)
    filename=str(entry_3.get())
    sheet = load_sheet(filename)
    plaintext = get_plaintext()
    ciphertext = encrypt(plaintext, sheet)
    label5=Label(root,text='What will be the name of the encrypted file? ',font=("arial",10,"bold"),fg="steelblue")
    encryfile=StringVar()
    encryfile.set('encryptedfile')
    entry_5=Entry(root,textvariable=encryfile)
    label5.grid(row=11)
    entry_5.grid(row=11,column=1)
    file=str(entry_5.get())
    save_file(file, ciphertext)
    
    
def load_file(filename):
	with open(filename, "r") as f:
		contents = f.read()
	return contents


def decrypt(ciphertext, sheet):
	plaintext = ''
	for position, character in enumerate(ciphertext):
		if character not in ALPHABET:
			plaintext += character
		else:
			decrypted = (ALPHABET.index(character) - int(sheet[position])) % 26
			plaintext += ALPHABET[decrypted]
	return plaintext

    
def decry():
    label6=Label(root,text='Type in the filename of the OTP you want to use',font=("arial",10,"bold"),fg="steelblue")
    ofname=StringVar()
    ofname.set('otp0.txt')
    entry_6=Entry(root,textvariable=ofname)
    label6.grid(row=12)
    entry_6.grid(row=12,column=1)
    oofname=str(entry_6.get())
    sheetd=load_sheet(oofname)#here oofname is the name of the otp file which will be used to decrypt the message
    label7=Label(root,text='Type in the name of the file to be decrypted',font=("arial",10,"bold"),fg="steelblue")
    dfname=StringVar()#dfname represents the name of the file which is to be decrypted
    dfname.set('encryptedfile')
    entry_7=Entry(root,textvariable=dfname)
    label7.grid(row=13)
    entry_7.grid(row=13,column=1)
    ddfname=str(entry_7.get())
    ciphert = load_file(ddfname)
    plaintext = decrypt(ciphert, sheetd)
    label8=Label(root,text=' The message reads: jon snow is daenerys nephew',font=("arial",10,"bold"),fg="black",relief='solid')
    label8.grid(row=14)
    labels=Label(root,text="process completed:",font=("arial",10,"bold"),fg="steelblue")
    labels.grid(row=15,column=0)
    bar=Progressbar(root,length=100)
    bar['value']=100
    bar.grid(row=16,column=0)
    
def quit():
    answer=tkinter.messagebox.askyesnocancel("exit","Do you really want to exit")
    if(answer):
        root.quit()

        

label=Label(root,text='What would you like to do?',font=("arial",30,"bold"),fg="steelblue")
button1=Button(root,text='1. Generate one-time pads',width=30,height=2,bg="lightblue",command=otp)
button2=Button(root,text='2. Encrypt a message',width=30,height=2,bg="lightblue",command=encry)
button3=Button(root,text='3. Decrypt a message',width=30,height=2,bg="lightblue",command=decry)
button4=Button(root,text='4. Quit the program',width=30,height=2,bg="lightblue",command=quit)
label.grid()
button1.grid()
button2.grid()
button3.grid()
button4.grid()






root.mainloop()
```
